```python
import alldsfsfuncs as me
import os
os.chdir(r"C:\Users\Barnett Sailcloth\Documents\Projects\Python\Data Science From Scratch")
print os.getcwd()


def predict(alpha, beta, x_i):
    return beta*x_i + alpha

def error(alpha, beta, x_i, y_i):
    return y_i - predict(alpha, beta, x_i, y_i)

def sum_of_squared_errors(alpha, beta, x, y):
    return sum(error(alpha, beta, x_i, y_i)**2 for x_i, y_i in zip(x,y))

def least_squares_fit(x, y):
    """given training values for x and y, find the least squares values of alpha and beta"""
    beta = me.correlation(x,y)*me.standard_deviation(y) / me.standard_deviation(x)
    alpha=me.mean(y) - beta*me.mean(x)
    return alpha, beta
        
```

lets make a multiple regression model:

1. what doe it look like?

linear regression deals with data for one explanatory variable $\bf{x}$ with response $\bf{y}$, where

\begin{equation}
\bf{y} = \alpha + \beta\bf{x} + \bf{\epsilon}
\end{equation}

and $\alpha$ and $\beta$ are parameters that are tuned to minimise $\bf{\epsilon} \cdot \bf{\epsilon}$.

Linear regression estimates a response variable $\bf{y}$ with a single explanatory variable $\bf{x}$
